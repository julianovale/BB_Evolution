# Equação da reta

$ax+by+c = 0$


Exemplo: $-1x_1 + 4x_2 + 0.4 $ 


```python
import numpy as np
import matplotlib.pyplot as plt

a = -1
b = 4
c = 0.4

def plotline(a, b, c):
  x = np.linspace(-2, 4, 50)
  y = (-a*x-c)/b
  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2, 4, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid()

plotline(a, b, c)
```

```python
p1 = (2, 0.4)
ret1 = a*p1[0] + b*p1[1] + c
p2 = (1, 0.6)
ret2 = a*p2[0] + b*p2[1] + c
p3 = (3, -0.4)
ret3 = a*p3[0] + b*p3[1] + c

plotline(a, b, c)
plt.plot(p1[0], p1[1], color='y', marker='o')
plt.plot(p2[0], p2[1], color='g', marker='o')
plt.plot(p3[0], p3[1], color='r', marker='o')
```

```python
print("p1 = %.1f" % ret1) # qualquer ponto sobre a reta, ao solucionar a equação para aquele ponto, o retorno é "0"
print("p2 = %.1f" % ret2) # qualquer ponto acima da reta, retorna um valor positivo
print("p3 = %.1f" % ret3) # qualquer ponto abaixo da reta, retorna um valor negativo
# quanto maior o módulo do valor, mais distante da reta
# então a reta pode ser usada como uma fronteira de decisão (um classificador)
```

### Então: 

$ax + by + c$ , quando falamos de redes neurais são $w_1x_1 + w_2x_2 + b$ onde: w são os pesos e b o viés (que só desloca a reta no espaço, sem alterar característica angular).

### Generalizando:

$$y = \sum_{i=1}^{d}w_ix_i+b$$
