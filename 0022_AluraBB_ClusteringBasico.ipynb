```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
caminho_arquivo = 'https://raw.githubusercontent.com/alura-cursos/Clustering/Aula-1.-An%C3%A1lise-Explorat%C3%B3ria/Wine.csv'

df = pd.read_csv(caminho_arquivo)
df.head()
```

```python
print("Número de amostras: ", df.shape[0], " - Número de atributos: ", df.shape[1])
```

```python
dicionario = {
    'Alcohol' : 'Alcool',
    'Ash' : 'Po',
    'Ash_Alcanity' : 'Alcalinidade_po',
    'Magnesium' : 'Magnesio',
    'Color_Intensity' : 'Intensidade_cor'
    }
df = df.rename(columns=(dicionario))
df.head()
```

```python
df.describe()
```

```python
matriz_corr = df.corr()
matriz_corr
```

```python
pip install biokit
```

```python
from biokit.viz import corrplot
import matplotlib.pyplot as plt
```

```python
corr_graph = corrplot.Corrplot(matriz_corr)
corr_graph.plot(upper='ellipse', fontsize = 'x-large')
fig = plt.gcf()
fig.set_size_inches(13,8)
fig.show()
```

```python
# interpretação da figura acima: o índice de correlação é intuitivo. 
# O formato da elipse demonstra o espalhamento dos dados.
# Então, quanto mais fino (uma linha) mais correlacionado
```

```python
# Normalização manual = MinMaxScaler
```

```python
df_01 = df.copy()
```

```python
atributos = df.columns
```

```python
for atributo in atributos:
  df_01[atributo] = (df_01[atributo]-min(df_01[atributo]))/(max(df_01[atributo] - min(df_01[atributo])))
```

```python
df_01.head()
```

```python
df_01.describe()
```

```python
# Normalização usando o Sklearn
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
np_df = min_max_scaler.fit_transform(df)
df_02 = pd.DataFrame(np_df, columns=atributos)
```

```python
df_02.head()
```

```python
df_02.describe()
```

```python
# Des-normalizando os dados
```

```python
np_df = min_max_scaler.inverse_transform(df_02)
df_03 = pd.DataFrame(np_df, columns=atributos)
```

```python
df_03.head()
```

```python
df_03.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
agrupador = KMeans(n_clusters = 4)
```

```python
agrupador.fit(df_02)
```

```python
labels = agrupador.labels_
print(labels)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_02['Intensidade_cor'], y = df_02['Alcool'],
                         mode='markers', 
                         marker=dict(color=agrupador.labels_.astype(np.float)), 
                         text = labels))
fig.show()
```

```python
agrupador = KMeans(n_clusters=3)
agrupador.fit(df_02)
labels = agrupador.labels_
labels
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_02['Intensidade_cor'], y = df_02['Alcool'],
                         mode='markers', 
                         marker=dict(color=agrupador.labels_.astype(np.float)), 
                         text = labels))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=df['Intensidade_cor'], y=df['Alcool'], z=df['Proline'], 
                           mode = 'markers',
                           marker = dict(color = labels.astype(np.float)),
                           text = labels))
fig.update_layout(scene=dict(xaxis_title='Intensidade de cor', yaxis_title='Alcool', zaxis_title='Proline'))
fig.show()
```
